```python
# Create a new dataframe
avg_deathrate_df = pd.DataFrame()

# YOUR CODE HERE

# Find average for each country
avg_deathrate_df ["Avg.Death Rate"] = cleaned_df.groupby("Country")["Death Rate"].mean()

# Set the index as a column - countries
avg_deathrate_df["Country"] = avg_deathrate_df .index

# Drop the index
avg_deathrate_df .reset_index(drop=True, inplace=True)

# Sort the rows based of Average Birth Rate
avg_deathrate_df .sort_values("Avg.Death Rate", inplace=True, ascending=False)

# Display
avg_deathrate_df .head()
```

## Making your own question 02? (1.5 points)

**Question:** How to predict female's life expectancy?

**Answering this question will** : Help the government and people, find out the way to improve women's lifetime.

**How we answer this question**: Build up a model to predict women's life expectancy on several factors.

### Pre-processing

```python
# YOUR CODE HERE (OPTION)
y = df_copy['Female life expectancy ']
features = ['Birth Rate', 'Death Rate', 'Female Employment in Industry(%)', 'Electric Power Consumption(kWH per capita)',
            'Renewable Energy Consumption (%)','Fossil Fuel Consumption (%)','Male life expectancy']
X = df_copy[features]
```

```python
X.describe()
```

### Analyzing to get the answer

**Build model**

```python
# YOUR CODE 
from sklearn.tree import DecisionTreeRegressor

# Define model. Specify a number for random_state to ensure same results each run
female_life_model = DecisionTreeRegressor(random_state=1)

# Fit model
female_life_model.fit(X, y)
```

```python
print("Making predictions for the following 5 houses:")
print(X.head())
print("The predictions are")
print(female_life_model.predict(X.head()))
```

**MODEL VALIDATION**

```python
from sklearn.metrics import mean_absolute_error

predicted_age = female_life_model.predict(X)
mean_absolute_error(y, predicted_age)
```

```python
from sklearn.model_selection import train_test_split

# split data into training and validation data, for both features and target
# The split is based on a random number generator. Supplying a numeric value to
# the random_state argument guarantees we get the same split every time we
# run this script.
train_X, val_X, train_y, val_y = train_test_split(X, y, random_state = 0)
# Define model
female_life_model = DecisionTreeRegressor()
# Fit model
female_life_model.fit(train_X, train_y)

# get predicted prices on validation data
val_predictions = female_life_model.predict(val_X)
print(mean_absolute_error(val_y, val_predictions))
```

## Making your own question 03? 

**Question:** How is the use of energy in the most developed countries?

**Answering this question will** : Help us have a general view about the impact these countries caused on the environment. Through this process we can learn the advanced from them, or call for the sake of the Earth.

**How we answer this question**: Collect data on `Renewable Energy Consumption (%), Fossil Fuel Consumption (%)` and draw bar chart for observing.

### Preprocessing

```python
# YOUR CODE HERE (OPTION)
energy_df =  pd.DataFrame(cleaned_df.groupby('Country')['GDP in USD',  'Renewable Energy Consumption (%)', 'Fossil Fuel Consumption (%)', 'Year'].mean())
energy_df = energy_df.sort_values(by = ['GDP in USD'], ascending = False)[:5]
energy_df['Country'] = energy_df.index
energy_df = energy_df.reset_index(drop=True)
energy_df 
```

```python
N = 5
 
renew = energy_df['Renewable Energy Consumption (%)']
fuel =  energy_df['Fossil Fuel Consumption (%)']

ind = np.arange(N)   
width = 0.35 
 
fig = plt.subplots(figsize =(10, 7))
p1 = plt.bar(ind, renew, width)
p2 = plt.bar(ind, fuel, width, bottom = renew)
 
plt.ylabel('Percentage')
plt.title('The consumption of energy of the 5 highest GDP countries in 2022')
plt.xticks(ind,  energy_df['Country'])
plt.yticks(np.arange(0, 81, 10))
plt.legend((p1[0], p2[0]), ('Renewable Energy Consumption (%)', 'Fossil Fuel Consumption (%)'))
 
plt.show()
```

### OBSERVATION ### 
    - Brazil is the only countries that consump a significant amount of renewable energy. Canada ranks the seconds with       about 20%.
    - United State, Mexico and Argentina, however, still depend on fossil fuel energy to develop the country.

## Making your own question 04?
